## Warm Up Exercise

Define a triangle class that has a list (named sides) as the only property, with the constructor that can intialize the three sides. Provide a method that calculates the areas of the triangle. 



````python
3 ** 3
````

````python
9 ** 0.5
````

````python
class Triangle:
    def __init__(self, a, b, c):
        ## Checking 
        self.__sides = [a, b, c]
    
    def area(self):
        print("Triangle area")
        
        a = self.__sides[0]
        b = self.__sides[1]
        c = self.__sides[2]
        
        s = (a + b + c)/2
        area = (s * (s-a) * (s-b) * (s-c)) ** 0.5
        
        return area
    
t = Triangle(3, 4, 5)
print(t.area())
        
    
````

## Inheritance

````python
class Animal:
    def __init__(self, name):
        self.name = name
        
    def say_name(self):
        print("My name is", self.name)
        
class Dog(Animal):
    def __init__(self, name):
        super().__init__(name)
        
    def say_name(self):
        print("I am a dog.")
        super().say_name()
        
class Cat(Animal):
    def __init__(self, name):
        super().__init__(name)
        
    def say_name(self):
        print("I am a cat.")
        super().say_name()
        
````

````python
d = Dog("Bob")
d.say_name()
````

````python
c = Cat("Alice")
c.say_name()
````

````python
class Farm:
    def __init__(self, animals):
        self.animals = animals
        
    def everyone_say_name(self):
        for animal in self.animals:
            animal.say_name()
````

````python
f = Farm([Cat("Alice"), Dog("Bob"), Animal("Cathy")])
````

````python
f.everyone_say_name()
````
